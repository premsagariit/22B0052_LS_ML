```python
def z_score(x_train):

  # ---------
  x_mean = np.mean(x_train, axis=0)
  x_std = np.std(x_train, axis=0)
  x_train = (x_train - x_mean)/x_std
  # ---------

  return x_train, x_std, x_mean
```

```python
def cost(x_train,y_train,w,b):

  # ---------
  m = x_train.shape[0]
  error = 0
  for i in range(m):
      f_wb_i = np.dot(x_train[i], w) + b
      error += (f_wb_i - y_train[i])**2
  loss = error/(2*m)
  # ---------

  return loss
```

```python
def gradient_descent(x_train,y_train, w, b, lr = 6.0e-3, num_iters = 1000):
  # ---------
  m, n = x_train.shape
  for iter in range(num_iters):
      dj_dw = np.zeros((n,))
      dj_db = 0.0

      for i in range(m):
          f_wb_i = np.dot(x_train[i], w) + b
          error = f_wb_i - y_train[i]
          for j in range(n):
              dj_dw[j] += error * x_train[i, j]
          dj_db += error

      # Average over all examples
      dj_dw /= m
      dj_db /= m

      # Update weights and bias
      w -= lr * dj_dw
      b -= lr * dj_db
  # ---------
  return w,b
```

```python
x_train = x_train.astype(np.float64)
x_train,x_std,x_mean = z_score(x_train)

np.random.seed(2147483647)
w = np.zeros((x_train.shape[1],))
b = 0.0

old_cost = 0

while abs(old_cost - cost(x_train,y_train,w,b)) > 0.00001:
  old_cost = cost(x_train,y_train,w,b)
  w,b = gradient_descent(x_train,y_train,w,b)

x_predict = pd.read_excel('drive/MyDrive/Learner Space/ML/Assignment_1/Test data.xlsx').iloc[:,:8].to_numpy()
x_predict = feature_changing(x_predict)
x_predict = (x_predict - x_mean)/x_std
ans = pd.read_excel('drive/MyDrive/Learner Space/ML/Assignment_1/Test data.xlsx').iloc[:,8].to_numpy()

y_predict = np.dot(x_predict,w) + b

accuracy = 0
for dim in range(len(ans)):
  if abs(y_predict[dim]-ans[dim])<0.5: # do not change the tolerance as you'll be checked on +- 0.5 error only
    accuracy += 1
accuracy = round(accuracy*100 / len(ans),2)
ok = 'Congratulations' if accuracy>95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
